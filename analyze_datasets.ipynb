```python
# Import libraries
import os
import pandas as pd
from IPython.display import display
```

```python
# Function to present the key characteristics of each dataset
def present_dataset(data_path, dataset):
    current_path = data_path + dataset
    print(f'\n{os.listdir(current_path)}\n')
    
    data = pd.read_csv(current_path + '/data.csv', index_col=0)
    print(f'First 5 rows of {dataset}:')
    display(data.head())    
    print(f'\nLength of {dataset}: {len(data)}\n')
    
    # see if these displays can be put inside a print statement
    agg_mat = pd.read_csv(current_path + '/agg_mat.csv', index_col=0)
    print('First 5 rows of agg_mat:') 
    display(agg_mat.head())
    print(f'\nLength of agg_mat: {len(agg_mat)}\n')
    print('Some nodes in the hierarchy:') 
    display(agg_mat.index.unique()[: 5])
    print('Proof that index contains all, and columns just agg nodes:') 
    display(agg_mat.index.difference(agg_mat.columns)[: 5])
    print('Another proof of the same:')
    display(agg_mat.columns.difference(agg_mat.index))
    
    tags = pd.DataFrame()
    try: 
        tags = pd.read_csv(current_path + '/tags.csv', header=None)
        print('Showing part of the hierarchy:')
        display(tags.iloc[: 5, :])
    except:
        pass
    
    return data, agg_mat, tags
```

```python
data_path = 'C:/Users/User/Desktop/Fakultet/2.Letnik/mlds_2/reproduction/RC_repo/experiments/data/'
```

## Displaying key characteristics of each dataset

```python
datasets = os.listdir(data_path)
for dataset in datasets:
    present_dataset(data_path, dataset)
```

**The folder, where the CSV files for a particular dataset are stored, differed from one dataset to another. So, in the continuation we investigate those differences to see if there is something unusual.**
